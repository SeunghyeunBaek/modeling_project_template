```python
import os
import sys

PROJECT_DIR = os.path.dirname(os.getcwd())
sys.path.append(PROJECT_DIR)

from modules.utils import load_json, load_pickle
import pandas as pd
from sklearn.metrics import classification_report
```

```python

predict_serial = '20210829_170539_20210829_175308'

predict_dir = os.path.join(PROJECT_DIR, 'results', 'predict', predict_serial)
gt = load_json(os.path.join(predict_dir, 'gt.json'))
pred = load_json(os.path.join(predict_dir, 'pred.json'))

filenames = list(gt.keys())
gts = list(gt.values())
preds = list(pred.values())

df = pd.DataFrame.from_dict({'filename': filenames, 'gt': gts, 'pred': preds})

clf_report = classification_report(gts, preds, digits=4, zero_division=False)
cm = pd.crosstab(df['gt'], df['pred'], margins=True)

print(clf_report)
display(cm)
```
